```python
#setting up imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sc
from matplotlib import cm
import matplotlib.colors
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
%matplotlib inline
```

# Part 1.

```python
#deffining the matrix A
A = np.array([[2,2,-2],[1,3,1],[1,2,2]])
```

# Part 2.

```python
#calculating the eigenvalues and respective eigen vectors for the matrix A
B = np.linalg.eig(A)

#rearanging the data stored  by the array B such that the eignvalues and eigenvectors each have a varriable assigned to them
#creating a variable for the eigenvalues
vals = B[0]
#creating a variable for all the eigenvectors
vecs = B[1]
#extracting the eigenvectors from the array
vecs1 = vecs[0]
vecs2 = vecs[1]
vecs3 = vecs[2]
vec1 = [vecs1[0],vecs2[0],vecs3[0]]
vec2 = [vecs1[1],vecs2[1],vecs3[1]]
vec3 = [vecs1[2],vecs2[2],vecs3[2]]
vec = [vec1,vec2,vec3]
```

```python
# using a loop to print the eigenvalues with their respective eigenvectors
n = 0
while n < 3:
    print("For the eigenvalue", vals[n], "the eigenvector is given by", vec[n])
    n = n + 1
```

# Part 3.

```python
#using a loop to compare the LHS and RHS of the characteristic eigenvector equation using the calculated eigenvectors
n = 0
while n < 3:
    lamx = np.dot(vals[n],vec[n])
    Ax = np.dot(A,vec[n])
    print("For the eigenvalue", vals[n], "A*X equals", Ax, "and lambda*X equals", lamx)
    n = n + 1
```

# Excercise 2

# Part 1.

```python
#creating 1000 random x and y velocities for the particles
rngx = np.random.default_rng().normal(loc=0.0,scale=1.0,size=1000)
rngy = np.random.default_rng().normal(loc=0.0,scale=1.0,size=1000)
```

```python
#using a loop to calculate the speed of the velocity of each particle
n = 0
v = [0]*1000
while n <1000:
    v[n] = (rngx[n]**2 + rngy[n]**2)**0.5
    n = n + 1
```

# Part 2.

```python
#plotting a histogram of the x-direction velocities
plt.style.use('_mpl-gallery')
figx, ax = plt.subplots()

ax.hist(rngx, bins=100, linewidth=0.5, color="red", edgecolor="blue")
#creating a suitably sized graph with gridlines
ax.set(xlim=(-4, 4), xticks=np.linspace(-4, 4, 9),
       ylim=(0, 50), yticks=np.linspace(0, 50, 11))

#labling the axis
plt.xlabel("x-velocities (m/s)")
plt.ylabel("quantity")
#displaying the histogram
plt.show()
```

```python
#plotting a histogram of the y-direction velocities
plt.style.use('_mpl-gallery')
figy, ax = plt.subplots()

ax.hist(rngy, bins=100, linewidth=0.5, color="red", edgecolor="blue")
#creating a suitably sized graph with gridlines
ax.set(xlim=(-4, 4), xticks=np.linspace(-4, 4, 9),
       ylim=(0, 50), yticks=np.linspace(0, 50, 11))

#labling the axis
plt.xlabel("y-velocities (m/s)")
plt.ylabel("quantity")
#displaying the histogram
plt.show()
```

```python
#plotting a histogram of the overall speed
plt.style.use('_mpl-gallery')
figv, ax = plt.subplots()

ax.hist(v, bins=100, linewidth=0.5, color="red", edgecolor="blue")
#creating a suitably sized graph with gridlines
ax.set(xlim=(0, 2), xticks=np.linspace(0, 4, 5),
       ylim=(0, 50), yticks=np.linspace(0, 50, 11))

#labling the axis
plt.xlabel("Overall speeds (m/s)")
plt.ylabel("quantity")
#displaying the histogram
plt.show()
```

# Part 3.

```python
#plotting a scattergraph of the y-velocities against their respective x-velocities
plt.style.use('_mpl-gallery')
figs, ax = plt.subplots()

ax.scatter(rngx, rngy, s=5, c="red", edgecolor = "blue")
#creating a suitably sized graph with gridlines
ax.set(xlim=(-4, 4), xticks=np.arange(-4, 4),
       ylim=(-4, 4), yticks=np.arange(-4, 4))

#labling the axis
plt.xlabel("x-velocities (m/s)")
plt.ylabel("y-velcoities (m/s)")
#displaying the scattergraph
plt.show()
```

```python
#plotting a colourmap with a colourbar of the y-velocities againt x-velocities with the colour representing the overall speed
#creating the colourbar parametres
cvals  = [ 0, 2]
colors = ["red", "blue"]

#creating the colourmap
norm=plt.Normalize(min(cvals),max(cvals))
tuples = list(zip(map(norm,cvals), colors))
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", tuples)

#plotting the colourmap
fig, ax = plt.subplots()
ax.scatter(rngx, rngy, c=v, cmap=cmap, norm=norm, s=5)
plt.scatter(rngx, rngy, c=v, cmap=cmap, norm=norm, s=5)
#creating a suitably sized graph with gridlines
ax.set(xlim=(-4, 4), xticks=np.arange(-4, 5),
       ylim=(-4, 4), yticks=np.arange(-4, 5))
#labling the axis
plt.xlabel("x-velocities (m/s)")
plt.ylabel("y-velocities (m/s)")
#plotting the colourbar
plt.colorbar()
#displaying the colourmap and respective colourbar
plt.show
```
